```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0 , 10 , 11)
y = np.linspace(0 , 10 , 11)
plt.plot(x,y,marker = 'o')
plt.title('First plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.grid()
plt.show()
plt.savefig('firstplot')

plt.close()
```
